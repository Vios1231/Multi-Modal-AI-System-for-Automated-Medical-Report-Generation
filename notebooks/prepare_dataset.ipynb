```python
import os
import xml.etree.ElementTree as ET
import pandas as pd

# Folder paths
images_dir = "data/images"
reports_dir = "data/reports"

print("Total gambar:", len(os.listdir(images_dir)))
print("Total report XML:", len(os.listdir(reports_dir)))

xml_files = [f for f in os.listdir(reports_dir) if f.endswith(".xml")]

data = []

for xml_file in xml_files:
    xml_path = os.path.join(reports_dir, xml_file)

    try:
        tree = ET.parse(xml_path)
        root = tree.getroot()

        # Ambil teks laporan (gabungkan semua <AbstractText>)
        report_text = " ".join([
            elem.text.strip()
            for elem in root.findall(".//AbstractText")
            if elem.text
        ])

        # Ambil nama file gambar terkait (id di <parentImage>)
        image_ids = [img.attrib["id"] for img in root.findall(".//parentImage")]

        for img_id in image_ids:
            image_filename = f"{img_id}.png"
            image_path = os.path.join(images_dir, image_filename)

            if os.path.exists(image_path):
                data.append({
                    "image": image_path,
                    "report": report_text
                })

    except ET.ParseError:
        print(f"⚠️ Error parsing {xml_file}")
        continue

# Buat DataFrame
df = pd.DataFrame(data)
print("Total pasangan gambar-report:", len(df))
df.head(5)
```

```python
df.to_csv("data/iu_xray_dataset.csv", index=False)
print("✅ Dataset final disimpan di data/iu_xray_dataset.csv")
```
